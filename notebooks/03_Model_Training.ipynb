```python
import os
HOME = os.getcwd()
HOME = HOME[0:HOME.find('notebooks')]
os.makedirs(HOME + 'models', exist_ok=True)
MODEL_FOLDER = HOME + 'models\\'
DATA_FOLDER = HOME + 'data\\'
```

```python
model = YOLO(MODEL_FOLDER + 'yolov8n.pt')
```

```python
import os
os.environ['PYTORCH_CUDA_ALLOC_CONF'] = 'expandable_segments:True'
```

```python
import torch
torch.cuda.empty_cache()
```

```python
model.train(
    data=DATA_FOLDER + 'data.yaml',
    epochs=300,
    imgsz=640,
    batch=16,
    device=0,
    project=MODEL_FOLDER,
    name='drowning_detection_yolov8',
    save_dir=MODEL_FOLDER,
    # cache=False
)
```
